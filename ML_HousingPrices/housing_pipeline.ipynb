```python
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
y = train_data.SalePrice
X = train_data.drop(['SalePrice'], axis=1)

X.head()
```

```python
y.head()
```

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
X_train_full.columns
```

```python
missing_val_count_by_column = (X_train_full.isnull().sum())
print(X_train_full.shape)
infs_val = (X_train_full.isin([np.inf,-np.inf]).sum())
print("\nMissing values:")
print(missing_val_count_by_column[missing_val_count_by_column > 0])
print("\nInfinity values:")
print(infs_val[infs_val > 0])
```

```python
categorical_cols = [col_name for col_name in X_train_full.columns if X_train_full[col_name].nunique() < 10 and 
                    X_train_full[col_name].dtype == 'object']
print(categorical_cols)
```

```python
num_cols = [col_name for col_name in X_train_full.columns if X_train_full[col_name].dtype in ['int64', 'float64']]
print(num_cols)
```

```python
my_cols = categorical_cols + num_cols
X_train = X_train_full[my_cols].copy()
X_valid = X_valid_full[my_cols].copy()
X_train.head()
```

```python
# for num data
numerical_transformer = SimpleImputer(strategy='constant')

# for cat data
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))                            
])

# bunde processing
preprocessor = ColumnTransformer(
    transformers=[
    ('num', numerical_transformer, num_cols),
    ('cat', categorical_transformer, categorical_cols)
    ])
```

```python
model = RandomForestRegressor(n_estimators=1000, random_state=0)
```

```python
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                              ('model', model)
                              ])
my_pipeline.fit(X_train, y_train)

preds = my_pipeline.predict(X_valid)

mae = mean_absolute_error(preds, y_valid)
print (mae)
```

```python
# XGBOOST
from xgboost import XGBRegressor

model_2 = XGBRegressor(n_estimators=1000, learning_rate=0.17, random_state=0)
my_pipeline_2 = Pipeline(steps= [('preprocessor', preprocessor),
                                 ('model', model_2)
                                ])

my_pipeline_2.fit(X_train, y_train)          
preds_2 = my_pipeline_2.predict(X_valid)
mae_2 = mean_absolute_error(preds_2, y_valid)

print("MAE: ", mae_2)
```
