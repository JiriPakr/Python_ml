```python

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    fn_model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    fn_model.fit(train_X, train_y)
    preds_val_fn = fn_model.predict(val_X)
    mae_fn = mean_absolute_error(val_y, preds_val_fn)
    return(mae_fn)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 550, 610, 1000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
