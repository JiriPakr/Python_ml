## DATA

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
features_all = np.array(train_data.columns)
features_all = np.delete(features_all, np.where(features_all == 'SalePrice'))
X = train_data[features]
X_all = train_data[features_all]
y = train_data.SalePrice
features_all 
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
train_X_all, val_X_all, train_y_all, val_y_all = train_test_split(X_all, y, random_state=1)
train_X.head()
```

## MODEL

```python
random_forrest_model = RandomForestRegressor(max_leaf_nodes=610)
random_forrest_model.fit(train_X, train_y)
rf_val_predictions = random_forrest_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)

print("Validation MAE for Random Forest Model: {:,.2f}".format(rf_val_mae))
```

```python
random_forrest_model_all = RandomForestRegressor(max_leaf_nodes=4000)
random_forrest_model_all.fit(train_X_all, train_y_all)
rf_val_predictions_all = random_forrest_model.predict(val_X_all)
rf_val_mae_all = mean_absolute_error(rf_val_predictions, val_y_all)

print("Validation MAE for Random Forest Model: {:,.2f}".format(rf_val_mae_all))
```

```python

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    fn_model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    fn_model.fit(train_X, train_y)
    preds_val_fn = fn_model.predict(val_X)
    mae_fn = mean_absolute_error(val_y, preds_val_fn)
    return(mae_fn)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 550, 610, 1000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
